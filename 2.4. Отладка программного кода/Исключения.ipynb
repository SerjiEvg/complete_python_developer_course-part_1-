# Исключения

## Исключения могут быть разными

В любой программе, особенно большой, могут возникать ошибки. Некоторые видит программа(синтаксичесикие), а некоторые может увидеть только сам программист (логические). 

Причин возникновения ошибок может быть множество. Например, программист не соблюдает правила синтаксиса языка Python. И в таком случае программа аварийно завершает свою работу, так как интерператор, встретив ошибочное выражение, не знает как его обработать. По итогу на экране программист видит сообщение об ошибке (traceback).

Кроме синтаксической ошибки существует и второй вид ошибок - семантические. Они возникают в следствие неправильного содержания действий и использования недопустимых значений величин.

В терминологии Python возникло такое понятие, как исключение. **Исключения -** это ошибки, обнаруженные при отладке программы. 

## Синтаксические ошибки или исключения

Тогда при возникновении синтаксической ошибки можно сказать, что возникло исключение типа SyntaxError. Однако синтаксические ошибки все-таки принято относить к ошибкам, нежели исключеням, так как такого вида ошибки невозможно обработать.

## Иерархичность исключений

Кроме этого исключения имеют свою иерархичность. Их не нужно запоминать, однако стоит знать наиболее часто используемые:


1.   ZeroDivisionError – ошибки, возникающие вследствие деления на ноль;

````python
num = int(input())

print(num / 0)
````

2.   ValueError – ошибка, которая возникает из-за передачи неправильного значения аргумента во встроенную операцию или функцию;



````python
str_exmpl = 'hello world'

int(str_exmpl)
````

3. TypeError  – ошибка, которая возникает, когда операция или функция применяется к объекту несоответствующего типа

````python
num_one = '1'
num_two = 2

print(num_one + num_two)
print(num_one ** num_two)
print(num_one / num_two)
print(num_one * num_two) # не ошибка
````

4. IndexError – ошибки, возникающие вследствие неверно указанного индекса (элемент,указанный вне границ списка);


````python
list = ['air', 'fire', 'water', 'earth']

print(list[-1])
print(list[5]) # ошибка
````

5. NameError – ошибка, возникающая вследствие обращения к необъявленной переменной;


````python
print(number * 3)
````

## Исключения нужно обрабатывать, чтобы не прерывать работу программы

Конечно, можно просто исправить эту ошибку. Но а если нужные данные мы запрашиваем у пользователя. Допустим, пользователь хочет открыть файл, которого не существует или не существует документа, указанного пользователем. Тогда, в любом случае нельзя аварийно прерывать работу программы. Однако, возникшее исключение можно обработать. Для обработки исключений используют следующую конструкцию:
```python
try:
  <ваш код>
except <название исключения (необязательный параметр)>:
  <инструкция, в случае возникновения исключения>
```
В блоке **try** указывается участок кода, в котором может возникнуть ошибка, но а в **except** то, что нужно сделать в случае возникновения исключения.

Рассмотрим данную конструкцию на примере:


````python
f = open("test.txt", "w")
# print(f.read())
try:
  print(f.read())
except:
  print('Невозможно открыть файл')
print('Пользователь можете продолжить работу')
````

Обратите внимание, что здесь мы пытаемся открыть несуществующий файл. В попытке чтения текстового файла мы получаем исключение и успешно обрабатываем ее. При этом, даже не указывая типа исключения. Однако это есть не всегда хорошо, так как таким образом мы не сможем понять какого рода исключение произошло в случае этого запуска.

Рассмортим следующий пример:


````python
try:
  index = int(input())
  number = int(input())

  colors = ['red', 'green', 'blue', 'black', 'yellow']

  print(colors[index])
  print(45 / number)
except IndexError:
  print('Такого элемента не существует')
except ZeroDivisionError:
  print('На ноль делить нельзя')
except ValueError:
  print('Вы ввели неверный формат данных')
````

В этом случае при введении нецелочисленных данных происходит обработка обработка исключения, равно и то же самое будет если мы введем индекс, которого нет в списке, а также элемент 0. И в любом случае можем понять из-за чего именно произошло исключение.

Если же есть необходимость, чтобы в любом случае произошло какое-либо действие, не важно произошло исключение или нет, можно воспользоваться конструкцией **finally**:

````python
colors_numbers = ['red', 5, 8, 9, 'green',2 , 0, -1, 'blue', 'black', 'yellow', 10]
degree = int(input())

for i in range(0, len(colors_numbers)):
  try:
    print(int(colors_numbers[i]) ** degree)
  except ValueError:
    print('Такая операция не присуща такому типу данных')
  finally:
    print('Были обработаны данные с элементом ', colors_numbers[i] , '\n')
````

## Итог

Обработка исключительных ситуаций играет важную роль в разработке, особенно в больших проектах. Программа должна предупреждать пользователя о возникших ошибках, но никак не останавливать свою работу и застывать в одном положении. Именно для этих ситуаций жизненно необходима обработка исключений.
