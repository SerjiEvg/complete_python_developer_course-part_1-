## Обычно используемые встроенные функции включают в себя max, min и round.
* Используется max для поиска наибольшего значения одного или нескольких значений.
* Используйте, min чтобы найти самое маленькое значение.
* Оба работают как с символьными строками, так и с числами.
    * «Больше» и «Меньше» используются (0-9, AZ, az) для сравнения символов.
    * Это означает, что:
        * 'a' меньше чем 'b'
        * 'A' меньше чем 'a'
        * '0' меньше чем 'a'
    * Это применяется для сортировки в алфавитном порядке.

```python
print(max(1, 2, 3))
print(min('a', 'b', 'c'))
print(min('a', 'A'))
```

# Форматирование строки

Мы часто будем воспроизводить команду вывода данных на экран, и порой, может быть непонятен формат вывода, по которому трудно разобрать, как и что выводит команда **print()**. Поэтому для решения этой проблемы были созданы средства форматирования. **Форматирование**, проще говоря, это изменение внешнего вида кода и его результатов, с целью улучшения читабельности кода и его выводящихся результатов. В языке Python есть множество таких функций, но разберем основные:

1.   end = '' - аргумент внутри одинарных кавычках определяет символ, которым заканчивается выполнение функции print(). По умолчанию это символ переноса строки «\n», то есть после вывода всех данных управление переходит на следующую строку, и следующий вывод функции print начинается с новой строки:





```python
print('Красный')
print('Зеленый')

print('Синий', end = ' ')
print('Желтый')

print('Белый')
```

2.   sep = ' ' - аргумент определяет, что будет стоять между переданными объектами, напрмер строками. По умолчанию это символ пробела

```python
print('One', 'Two', 'Three')
print('One', 'Two', 'Three', sep = ' / ')
```

3. f-*строки* - позволяет строить длинные конструкции из комбинаций строк и имён переменных, путем вставки переменных в любое место при помощи конструкции {}, внутри которых укзывается имя переменной

```python
name = 'Daniel'
age = 24

print(f'My name is {name}. I am {age} years old')
```
