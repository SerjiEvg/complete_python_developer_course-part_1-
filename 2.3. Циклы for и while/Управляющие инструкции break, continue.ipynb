## Используйте инструкцию break для выхода из цикла
Довольно часто можно столкнуться с ситуацией, когда необходимо произвести выход из цикла. Например, когда проверяется результат и он не удовлетворяет каким-либо требованиям, тогда можно воспользоваться конструкцией break

```python
i = 0
while i >= 0:
  if i < 0:
    break
  i = int(input())
else:
  print('Было обнаружено отрицательное число: ', i)
```

Обратите внимание, что написанный код после инструкции break, не будет выполняться. Поэтому после ввода отрицательного числа, компилятор не запрашивает данные у пользователя, т.к уже произошел выход из тела цикла и управление передалось команде else для вывода соответствующего сообщения.

Заметка! Как и в случае с циклом for, с while также можно использовать конструкцию while ... else

## Используйте конструкцию continue для того, чтобы пропустить итерацию, не выходя из цикла
Для того, чтобы не выходить из цикла, но пропустить итерацию можно использовать инструкцию continue:

```python
i = 1

while i != 0:
  i = int(input())
  continue
```

**Важно!** Если использовать конструкции **break** и **continue** вне тела цикла, то это приведет к ошибке

## Бесконечные циклы не остановятся никогда

Из названия таких циклов уже понятно, что такие циклы не останавливаются никогда. Однако, это происходит в связи с тем, что цикл не имеет условия выхода. Например, 


```python
while True:
  print('Цикл работает')
```

Здесь видно, что цикл всегда будет выводить одно и то же сообщение бесконечное количество раз. Однако, если мы заменим эту программу на следующую

```python
while False:
  print('Цикл работает')
```

то цикл не будет работать и не выведет ни одно сообщение. Это происходит из-за того, что условие верное (True), соответственно происходит вход и выполнение тела цикла, но в этом случае не предусмотрено условие выхода из цикла. 
Другим примером бесконечных циклов является следующий:

```python
i = 1
while i < 10:
  print('Число цикла: ', i)
```

На самом деле проблему с бесконечными циклами можно решить путем простого добавления условия, обеспечивающего выход из цикла. Например для первого, можно сделать следующее:

```python
check = True
i = 0

while check == True:
  i += 1
  if i == 5: check = False
  else: check = True
  print(i)

  
```

а для второго добавить увеличение счетчика i:

```python
i = 0
while i < 10:
  print('Число цикла: ', i)
  i += 1
```

## Вложенные циклы, как способ выполнения одних и тех же действий множество раз

До этого, нами были использованы циклы, действия в которых, выполняются единожды для одних парметров. Если же нам нужно выполнить один и тот же участок кода множество раз, можно использовать вложенные циклы, то есть внутри одного цикла будет работатать другой цикл. Довольно часто вложенные циклы используются для вывода матрицы:

```python

for i in range(-10, 0):
  for j in range(2, 12):
    print(i + j, end = ' \t')
  print()
```

Или, например для вывода элементов матрицы. При этом можно использовать функцию while. Например, есть некоторый двумерный массив:




```python
a = [[2, 5], [9, 0], [6, 1]]
i = j = 0
while i < len(a):
  while j < len(a[i]):
    print(a[i][j], end = ' ')
    j += 1
  print()
  i += 1
  j = 0
```

Но легче использовать для этих целей цикл for, так как перебор по элементам матрицы проще реализовать именно с помощью этого цикла

```python
a = [[2, 5], [9, 0], [6, 1]]

for i in range(len(a)):
  for j in range(len(a[i])):
    print(a[i][j], end = ' ')
  print()
```

Обратите внимание на то, что цикл может иметь множественную степень вложенности, то есть при необходимости во внешний цикл можно вложить еще ннесколько. Однако такое не приветствуется, так как с углублением цикла расчеты будут увеличиваться, что здорово снизит скорость работы вашей программы.

## Инструкции break и continue находящиеся внутри вложенного цикла будут работать только внутри того цикла, в тело которого они были помещены

Если вы будете использовать конструкции break или continue во вложенных списках во внутреннем списке, то нужно знать, что выход из цикла или пропуск итерации будет тоже во внутреннем цикле. Например:


```python
num = int(input('Введите число: '))
for i in range(2, num):
  for j in range(2, i):
    if i % j == 0:
      print(i, 'нацело делится на ', j, ' = ', i//j)
      break
    else:
      print(i, 'простое число и не делится нацело на ', j)
```

Обратите внимание, что инструкция break находится во внутреннем цикле, соответственно, когда управление передается break, то команда исполняет выход из внутреннего цикла, но никак не из внешнего. В случае с continue правило то же самое:

```python
for i in range(0, 5):
  for j in range(-2, 2):
    if i * j == 0:
      continue
    print(i * j, end = ' ')
  print()
```

В этой программе вычисляется произведение двух чисел. Если итоговый результат умножения двух чисел равен 0, то происходит пропуск итерации и на экран не выводится результат 0.
