## Создание множества.

Множество в Python обозначается фигурной скобкой { }, причем Python автоматически удаляет повторяющиеся элементы.


```python
set1 = {"pop", "rock", "soul", "hard rock", "rock", "R&B", "rock", "disco"}
set1
```

Множество сожно создать на основе существующего списка:

```python

album_list = [ "Michael Jackson", "Thriller", 1982, "00:42:19", \
              "Pop, Rock, R&B", 46.0, 65, "30-Nov-82", None, 10.0]
album_set = set(album_list)             
album_set
```

```python
music_genres = set(["pop", "pop", "rock", "folk rock", "hard rock", "soul", \
                    "progressive rock", "soft rock", "R&B", "disco"])
music_genres
```

## Операции над множествами.

Некоторые операции над множествами используются для изменения набора значений. Создадим множество `A`

```python
A = set(["Thriller", "Back in Black", "AC/DC"])
A
```

C помощью метода `add()` можно добавить новый элемент в множество.

```python
A.add("NSYNC")
A
```

Если попытаться добавить этот же элемент еще раз, то ничего не произойдет. Множество - не любит копии.

```python
A.add("NSYNC")
A
```

С помощью метода remove() можно удалить элемент из множества.

```python
A.remove("NSYNC")
A
```

Также можно проверить на вхождение элемента в множество значений.

```python
"AC/DC" in A
```

## Логические операции над множествами.

Создадим два множества

```python
album_set1 = set(["Thriller", 'AC/DC', 'Back in Black'])
album_set2 = set([ "AC/DC", "Back in Black", "The Dark Side of the Moon"])
```

```python
album_set1, album_set2
```

Оба множества содержат AC/DC и Back in Black и представляются как пересечение на диаграмме. Операция пересечения множеств реализуется с момощью оператора `&`

```python
intersection = album_set1 & album_set2
intersection
```

Вы можете найти все элементы, которые содержатся в `album_set1`, используя метод `difference`.

```python
album_set1.difference(album_set2) 
```

Отображаются элементы, которые содержаться в множестве album_set1, но не включены во второе множество. И обратная операция

```python
album_set2.difference(album_set1) 
```

Объединение элементов множеств выполняется с помощью метода union().

```python
album_set1.union(album_set2)
```

Вы можете проверить, является ли набор надмножеством или подмножеством другого набора

```python
set(album_set1).issuperset(album_set2)   
```

```python
set(album_set2).issubset(album_set1)     
```

Вот пример, когда issubset() и issuperset() возвращает true:

```python
set({"Back in Black", "AC/DC"}).issubset(album_set1) 
```

```python
album_set1.issuperset({"Back in Black", "AC/DC"})   
```
