# Что такое консоль?

В широком смысле `Console` на английском означает «пульт управления». Игровая консоль – тут все понятно, это игровое устройство, подсоединяемое к компьютеру. Аналогично называют весь комплект устройств ввода-вывода: мышь, дисплей, клавиатура, а также внешние микрофоны и камеры.

`Интерфейс командной строки`, то есть окно для вывода системных сообщений и приема команд, – тоже «консоль».

Рассматривая программирование на Python понятие консоль будет употребляться в значении интерактивного инструмента для выполнения команд Python и отображения результатов. 

`Интерактивная консоль` предоставляет доступ к истории команд, всем встроенным функциям и установленным модулям Python. Она позволяет использовать автозаполнение, исследовать возможности Python и вставлять код в файлы программирования после проверки.

Эмуляцию консоли Python можно увидеть на главной странице [python.org](https://www.python.org/) и даже выполнить команды языка. Но эта консоль имеет ограниченный функционал и пользоваться ей для разработки программ не рекомендуется.

Доступ к интерактивной консоли Python можно получить с любого локального компьютера или сервера, на котором установлен Python. Например, если вы работаете в Windows, то вызовите командную строку windows, набрав `cmd` и в окне командной строки наберите `python`. 

Должна открыться интерактивная консоль Python, например, такая

Выйти из режима работы с консолью можно, нажав сочетание клавиш `Ctrl + D`. 

Также можно завершить работу с консолью более мягким способом, набрав команду `exit()` или по сочетанию клавиш `Ctrl + Z`.

# Вывод информации на консоль

## Вывод информации с помощью print()

Один из самых распространенных способов вывести данные в Python – это напечатать их в консоли. 

Если вы находитесь на этапе изучения языка, такой способ является основным для того, чтобы быстро просмотреть результат своей работы. Для вывода данных в консоль используется функция `print`.

Рассмотрим основные способы использования данной функции.

```python
print("Hello")
```

```python
print("Hello, " + "world!")
```

```python
print("Age: " + str(23))
```

По умолчанию, для разделения элементов в функции print используется пробел.


```python
print("A", "B", "C")
```

Для замены разделителя необходимо использовать параметр `sep` функции print.

```python
print("A", "B", "C", sep="#")
```

В качестве конечного элемента выводимой строки, используется символ перевода строки. 

```python
for i in range(3):
    print("i: " + str(i)) 
```

Для его замены используется параметр `end`.

```python
for i in range(3):
    print("[i: " + str(i) + "]", end=" -- ")
```

## Ввод данных из файла

Чтение данных из файла осуществляется с помощью методов `read(размер)` и `readline()`.

Метод `read(размер)` считывает из файла определенное количество символов, переданное в качестве аргумента. Если использовать этот метод без аргументов, то будет считан весь файл.

```python
f = open("test.txt", "r")
f.read()
```

```python
f.close()
```

В качестве аргумента метода можно передать количество символом, которое нужно считать.

```python
f = open("test.txt", "r")
f.read(7)
```

```python
f.close()
```

Метод readline() позволяет считать строку из открытого файла.

```python
f = open("test.txt", "r")
f.readline()
```

```python
f.close()
```

## Запись данных в файл

Для записи данных файл используется метод write(строка), при успешной записи он вернет количество записанных символов.

```python
f = open("test_out.txt", "a")
f.write("Test string")
```

```python
f.close()
```
